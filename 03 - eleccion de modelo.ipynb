```python
import pickle
ruta_pickle = 'data.pickle'  # Ruta al archivo data.pickle
with open(ruta_pickle, 'rb') as file:
    data = pickle.load(file)
d = pd.DataFrame(data)
d = d.head(10000)
```

##Elección de modelo

Esta parte del código toma 27 minutos en ejecutar

```python
X = d.drop('HasDetections', axis=1)  # Reemplaza 'variable_objetivo' con el nombre de tu variable objetivo
y = d['HasDetections'].astype(int)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear los clasificadores
svc = SVC(probability=True, random_state=42)
rf = RandomForestClassifier(random_state=42)
dt = DecisionTreeClassifier(random_state=42)

# Entrenar los clasificadores
svc.fit(X_train, y_train)
rf.fit(X_train, y_train)
dt.fit(X_train, y_train)

# Obtener las predicciones de probabilidad para los conjuntos de prueba
svc_probs = svc.predict_proba(X_test)[:, 1]
rf_probs = rf.predict_proba(X_test)[:, 1]
dt_probs = dt.predict_proba(X_test)[:, 1]

# Calcular el ROC AUC para cada modelo
svc_roc_auc = roc_auc_score(y_test, svc_probs)
rf_roc_auc = roc_auc_score(y_test, rf_probs)
dt_roc_auc = roc_auc_score(y_test, dt_probs)

# Imprimir los resultados
print("ROC AUC - SVC: {:.4f}".format(svc_roc_auc))
print("ROC AUC - Random Forest: {:.4f}".format(rf_roc_auc))
print("ROC AUC - Decision Tree: {:.4f}".format(dt_roc_auc))

# Comparar y seleccionar el modelo con el mejor ROC AUC
best_model = max([(svc_roc_auc, 'SVC'), (rf_roc_auc, 'Random Forest'), (dt_roc_auc, 'Decision Tree')], key=lambda x: x[0])
print("\nEl mejor modelo es: {}".format(best_model[1]))
```
