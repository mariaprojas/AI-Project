```python
for c in d.columns:
    print ("%20s"%c, d[c].dtype)
```

```python
d = d[[i for i in d.columns if not i in d._get_numeric_data()]]
```

```python
d = d.drop('MachineIdentifier', axis=1)
```

```python
ohd_list = []
for col in d.columns:
    
    # Codificación One-Hot de la columna
    ohd = pd.get_dummies(d[col], prefix=col, dummy_na=True)

    # agregar el DataFrame de codificación One-Hot a la lista
    ohd_list.append(ohd)

# Unir todos los DataFrames de codificación One-Hot en uno solo
d = pd.concat([d] + ohd_list, axis=1)

# Eliminar los DataFrames de codificación One-Hot individuales de la memoria
del ohd_list

d.head()
```

```python
from scipy.stats import ttest_ind
import random
d1 = d.copy()
d1 = d1.fillna(method="ffill")
d2 = d.copy()
d2 = d2.fillna(method="bfill")
d3 = d.copy()
# reemplaza los valores faltantes con la moda en cada columna
for col in d3.columns:
   mode = d3[col].value_counts().index[0]
   d3[col].fillna(mode, inplace=True)
d4 = d.copy()
for col in d4.columns:
   uni = d4[col].unique()
   d4[col].fillna(random.choice(list(uni)), inplace=True)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
# calcular el MSE
mse_d1 = mean_squared_error(d, d1)
print('MSE:', mse_d1)

# calcular el R²
r2_d1 = r2_score(d, d2)
print('R²:', r2_d1)

# calcular el MSE
mse_d2 = mean_squared_error(d, d2)
print('MSE:', mse_d2)

# calcular el R²
r2_d2 = r2_score(d, d2)
print('R²:', r2_d2)

# calcular el MSE
mse_d3 = mean_squared_error(d, d3)
print('MSE:', mse_d3)

# calcular el R²
r2_d3 = r2_score(d, d3)
print('R²:', r2_d3)

# calcular el MSE
mse_d4 = mean_squared_error(d, d4)
print('MSE:', mse_d4)

# calcular el R²
r2_d4 = r2_score(d, d4)
print('R²:', r2_d4                                                                        )
```
