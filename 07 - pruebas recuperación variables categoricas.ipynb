```python
d = d[[i for i in d.columns if not i in d._get_numeric_data()]]
```

```python
from scipy.stats import ttest_ind
import random
d1 = d.copy()
d1 = d1.fillna(method="ffill")
d1 = d1.fillna(method="bfill")
d2 = d.copy()
d2 = d2.fillna(method="bfill")
d2 = d2.fillna(method="ffill")
d3 = d.copy()
# reemplaza los valores faltantes con la moda en cada columna
for col in d3.columns:
   mode = d3[col].value_counts().index[0]
   d3[col].fillna(mode, inplace=True)
d4 = d.copy()
for col in d4.columns:
   uni = d4[col].unique()
   d4[col].fillna(random.choice(list(uni)), inplace=True)
```

```python
for col in d.columns:
  # Codificación One-Hot de la columna en cada dataframe
     ohd = pd.get_dummies(d[col], prefix=col) 
     del d[col]
     d= pd.concat([d, ohd], axis=1)
d.head()
```

Esta parte del código se demora alrededor de 25 minutos en ejecutar

```python
from scipy.stats import anderson_ksamp

n_samples = 1000
dfs = [d1.head(n_samples), d2.head(n_samples), d3.head(n_samples), d4.head(n_samples)]
difmeant = []
difstdt = []

for i, df in enumerate(dfs):
    df_mean = df.mean()
    df_std = df.std()
    d_mean = d.head(n_samples).mean()
    d_std = d.head(n_samples).std()

    difmean = np.mean(np.abs(df_mean - d_mean))
    difstd = np.mean(np.abs(df_std - d_std))
    difmeant.append(difmean)
    difstdt.append(difstd)

    coef = (pd.concat([df, d.head(n_samples)], axis=1).corr()).mean().mean()
    print(f"Difmean para el método {i+1}: {difmean}. Difstd para el método {i+1}: {difstd}. Coefcorr para el método {i+1}: {coef}.")
```
