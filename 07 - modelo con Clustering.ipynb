```python
import pickle
ruta_pickle = 'data.pickle'  # Ruta al archivo data.pickle
with open(ruta_pickle, 'rb') as file:
    data = pickle.load(file)
d = pd.DataFrame(data)
```

##Clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X= d.drop('HasDetections', axis=1)  
y= d['HasDetections'].astype(int)
n_clusters = 2

km = KMeans(n_clusters=n_clusters)
km.fit(X)
labels = km.fit_predict(X)
```

##Silhouette

Se evalua el clustering a través de Silhouette ya que el área bajo la curva ROC no es adecuada para evaluar el desempeño del clustering. 

*Esta parte del código se demora una hora ejecutando.*

```python
silhouette_avg = silhouette_score(X, labels)
print("La silueta media es:", silhouette_avg)
```
