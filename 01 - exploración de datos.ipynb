Variables categoricas

```python
ccols = [i for i in d.columns if not i in d._get_numeric_data()]
print (ccols)
```

```python
import dask.dataframe as dd

# Convertir el dataframe de Pandas en un objeto de Dask
df_dask = dd.from_pandas(d, npartitions=4)

# Lista de nombres de columnas categóricas
#ccols
# Contar los valores únicos en cada columna categórica en paralelo
for c in ccols:
    unique_values = df_dask[c].value_counts().compute()
    print(f"Columna {c} tiene {unique_values.shape[0]} valores únicos")
    print(unique_values)
```

###Valores faltantes

```python
k = d.isna().sum()
k1 = k[k!=0]
fig, ax = plt.subplots()
ax.plot(k1)
plt.xticks(rotation=90)
plt.xticks(fontsize=5)
plt.plot(k1)
```

###Distribución de la variable objetivo

```python
sns.distplot(d['HasDetections']);
plt.xticks([0, 1],["No", "Yes"])
```

### Distribución de otras variables

```python
c="Processor"
d[c].value_counts()
```

```python
plt.figure(figsize=(20,8))
for i,c in enumerate(["Census_ProcessorClass", "Platform", "Processor", "Census_GenuineStateName"]):    
    plt.subplot(2,4,i+1)
    k=d[[c,"HasDetections"]].dropna()
    for v in k[c].unique():
        sns.distplot(k.loc[k[c] == v, "HasDetections"], label=v)
        plt.title(c)
    plt.yticks([])
    plt.legend()

    plt.subplot(2, 4, i+5)
    vc = k[c].value_counts()
    sns.barplot(x=vc.index, y=vc.values)
    plt.xticks(rotation=90)   
```

### Correlación entre variables
